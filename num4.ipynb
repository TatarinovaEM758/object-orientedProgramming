```python
from datetime import datetime

class Home:
  def __init__(self, num_house, num_room, year_const, now_year):
    self.num_house = num_house
    self.num_room = num_room
    self.year_const = year_const
    self.now_year = now_year
  def quality(self):
    Q = self.num_room + 2 * (self.now_year - self.year_const)
    return Q

class Dist(Home):
  def __init__(self, num_house, num_room, year_const, now_year, district):
    Home.__init__(self, num_house, num_room, year_const, now_year)
    self.district = district
  def quality(self):
    if self.district == '':
      Qp = 2 * Home.quality(self)
    else:
      Qp = 0.5 * Home.quality(self)
    return Qp


try:
  now_year = datetime.now().year
  def hallo(obj):
    print("  {} = {}".format(obj.num_house, obj.quality()))
  a = Home(12, 3, 1956, now_year)
  b = Dist(11, 3, 1970, now_year, '')
  c = Dist(9, 4, 1950, now_year, '')
  hallo(a)
  hallo(b)
  hallo(c)
except TypeError:
  print("   !")
```
